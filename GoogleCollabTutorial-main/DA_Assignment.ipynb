# **(Question 2, Part 1)**

#  Drop the null values from the dataset. Use (dropna()) method.

# First we import Pandas Library to perform the operations given in Question 2

```python
import pandas as pd
```

# Reading the file

# CSV File

```python
df_csv = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/AirQualityUCI.csv")
df_csv
```

# Excel File

```python
df_xlsx = pd.read_excel("/content/drive/MyDrive/Data Sets for Data Visualisation/AirQualityUCI.xlsx")
df_xlsx
```

```python
# Print null values
print("Null values in the DataFrame:")
df_csv.isnull()
```

```python
# Print null values
print("Null values in the DataFrame:")
df_xlsx.isnull()
```

# We can see that only the CSV File contains Null Values, so dropping the null values from the csv file using dropna

```python
# Drop null values
df_csv.dropna(inplace=True)

# Print the DataFrame after dropping null values
df_csv
```

# **(Question 2, Part 2)**

# Replace NULL values with the number 130.




```python
# Replace null values with the number 130
df_csv.fillna(130, inplace=True)

# Print the DataFrame after replacing null values
df_csv
```

# **(Question 2, Part 3)**

# Filter the value of NO2(GT) > 200. Use data frame(df. loc) methods

Note: Used NO2(GT) as SO2 isn't available and value as 200 because no coulmn has value

```python
# Filter rows where 'NO2(GT)' > 100
filtered_df = df_xlsx.loc[df_xlsx['NO2(GT)'] > 200]

# Print the filtered DataFrame
filtered_df
```
