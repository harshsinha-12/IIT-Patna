```python
import pandas as pd
import matplotlib.pyplot as plt

a = pd.DataFrame(data['day'].value_counts())
a.reset_index(inplace = True)
plt.pie(a['day'],labels = a['index'], autopct = '%1.2f', explode = [0.1, 0, 0, 0])
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Assuming 'data' is your DataFrame
a = pd.DataFrame(data['day'].value_counts())
a.reset_index(inplace=True)

# Create a pie chart using Plotly Express
fig = px.pie(a, names='index', values='day', title='Distribution of Days')

# Customize the appearance if needed
fig.update_traces(textinfo='percent+label', pull=[0.1, 0, 0, 0])

# Show the chart
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is your DataFrame
a = pd.DataFrame(data['day'].value_counts())
a.reset_index(inplace=True)

# Create a pie chart using Seaborn
plt.figure(figsize=(6, 6))
sns.set(style="whitegrid")
sns.color_palette("pastel")
plt.pie(a['day'], labels=a['index'], autopct='%1.2f%%', startangle=140)

# Optional: Add a title to your pie chart
plt.title("Distribution of Days")

plt.show()
```
