Horizontal Stack Plot

```python
import matplotlib.pyplot as plt

# Data for the stack plot
categories = ['Category A', 'Category B', 'Category C']
values1 = [10, 15, 5]
values2 = [5, 20, 10]
values3 = [15, 10, 15]

# Create the stack plot
plt.figure(figsize=(8, 4))  # Set the figure size
plt.barh(categories, values1, label='Value 1')
plt.barh(categories, values2, left=values1, label='Value 2')
plt.barh(categories, values3, left=[v1 + v2 for v1, v2 in zip(values1, values2)], label='Value 3')

# Add labels and legend
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Stack Plot')
plt.legend()

# Show the plot
plt.show()
```

```python

import matplotlib.pyplot as plt
y1 = [31.80,54.17,47.38,47.62,68.40,52.83,58.56,36.65,92.38,76.84,133.56,59.86,47.88,43.00,27.45,55.63]
x1 = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023]

plt.bar(x1,y1, color = 'b', label = "Average", width = 0.5)

plt.plot()

plt.xlabel('Year')                                # xlabel and ylabel both are keywords
plt.ylabel('Average')

plt.legend()
plt.title('Virat Kohli ODI Average over the years')
plt.show()
```

```python
import plotly.express as px

y1 = [31.80, 54.17, 47.38, 47.62, 68.40, 52.83, 58.56, 36.65, 92.38, 76.84, 133.56, 59.86, 47.88, 43.00, 27.45, 55.63]
x1 = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

# Create a DataFrame with the data
import pandas as pd
df = pd.DataFrame({'Year': x1, 'Average': y1})

# Create a bar chart using Plotly Express
fig = px.bar(df, x='Year', y='Average', color_discrete_sequence=['blue'])

# Customize the chart layout
fig.update_layout(
    title='Virat Kohli ODI Average over the years',
    xaxis_title='Year',
    yaxis_title='Average'
)

# Show the chart
fig.show()
```
