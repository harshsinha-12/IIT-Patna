```python
import matplotlib.pyplot as plt
import numpy as np


# Generate 1000 random data plots from a normal distribution
data = np.random.randn(1000)

# Plot histogram
plt.hist(data, bins = 100, edgecolor = 'black') # Bins determines the number of data
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.xticks(rotation = 45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import re
from collections import Counter

# Generate 1000 random data plots from a normal distribution
data = np.random.randn(1000)
def create_hist(text):
  tokens = re.findall(r'\b\w+\b', text.lower())

  words_counts = Counter(tokens)
  # Plot histogram
  plt.figure(figsize= (10,5))
  plt.bar(words_counts.keys(), words_counts.values())
  #plt.hist(data, bins = 100, edgecolor = 'black') # Bins determines the number of data
  plt.xlabel('Value')
  plt.ylabel('Frequency')
  plt.title('Histogram of Random Data')
  plt.xticks(rotation = 45)
  plt.show()


text = "This is an example of histogram representation for text. The choice of data representation depends on Histogram"
create_hist(text)
```
