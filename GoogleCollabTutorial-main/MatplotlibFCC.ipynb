# Bar Chart

```python
labels = ['A', 'B', "C"]
values = [1,4,2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```

```python
labels = ['A', 'B', "C"]
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', '*', 'o']
for bar in bars:
  bar.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```

```python
fifa['Weight'] = [x.strip('lbs') if type(x) == str else x for x in fifa['Weight']]
```

```python
fifa['Weight']
```

```python
fifa['Weight'][0]
```

```python
fifa['Weight'] = [int(x.strip('lbs')) if type(x) == str else x for x in fifa['Weight']]
```

```python
fifa['Weight'][0]
```

```python
light = fifa.loc[fifa['Weight'] < 125].count()[0]
light_med = fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] < 150) ].count()[0]
med = fifa.loc[(fifa['Weight'] >= 150) & (fifa['Weight'] < 175) ].count()[0]
med_heavy = fifa.loc[(fifa['Weight'] >= 175) & (fifa['Weight'] < 200) ].count()[0]
heavy = fifa.loc[fifa['Weight'] > 200].count()[0]
```

```python
med_heavy
```

```python
weights = [light,light_med,med,med_heavy,heavy]
labels = ['Under 125', '125-150', '150,175','175-200','Over 200']
ex = [0,0,0,0.1,0]
plt.pie(weights, labels = labels,startangle = 30, shadow = True,
        explode = ex)
plt.show()
```

```python
weights = [light,light_med,med,med_heavy,heavy]
labels = ['Under 125', '125-150', '150-175','175-200','Over 200']
ex = [.4,.1,0,0,.4]
plt.style.use('ggplot')
plt.figure(figsize=(8,8))
plt.pie(weights, labels = labels,startangle = 30, shadow = True,autopct = '%.2f %%',
        pctdistance = 0.8, explode = ex)
plt.title('Weight Distribution of FIFA Players (in lbs)')
plt.show()
```
