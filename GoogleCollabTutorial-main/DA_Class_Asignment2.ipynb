# Rename a particular column using pandas dataframe.

```python
import pandas as pd

# Create a sample DataFrame
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
print(df)
# Rename column 'B' to 'New_Column_Name'
df = df.rename(columns={'B': 'New_Column_Name'})

# Display the DataFrame with the renamed column
print(df)
```

# Sort an index of dataframe using pandas library.

```python
import pandas as pd

# Create a sample DataFrame with a custom index
data = {'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10]}
index = ['C', 'A', 'E', 'B', 'D']
df = pd.DataFrame(data, index=index)
print(df)
print()
# Sort the index in ascending order
df_sorted = df.sort_index()

# Display the DataFrame with the sorted index
print(df_sorted)
```

```python
df_sorted_desc = df.sort_index(ascending=False)
print(df_sorted_desc)
```

# Use iloc function for pandas dataframe to select any particular column.

```python
import pandas as pd

# Create a sample DataFrame
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)

# Select the second column (column at index 1) using iloc
selected_column = df.iloc[:, 1]

# Display the selected column
print(selected_column)
```

# Create a sample and perform Group by operation on a data frame.

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value': [10, 20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)

# Perform a group-by operation based on the 'Category' column
grouped = df.groupby('Category')

# Calculate the sum of 'Value' within each group
result = grouped['Value'].sum()

# Display the grouped and aggregated data
print(result)
```

# Using dropna method to drop null values from pandas dataframe.

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with some null values
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [6, np.nan, 8, 9, 10],
    'C': [11, 12, 13, np.nan, 15]
}
df = pd.DataFrame(data)

# Remove rows with null values
df_dropped_rows = df.dropna()

# Remove columns with null values
df_dropped_columns = df.dropna(axis=1)

# Display the DataFrames with null values removed
print("DataFrame with rows containing null values dropped:")
print(df_dropped_rows)

print("\nDataFrame with columns containing null values dropped:")
print(df_dropped_columns)
```
