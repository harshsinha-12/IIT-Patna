# Merging 12 months of sales data into a single CSV File

```python
df_april = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_April_2019.csv")
df_aug = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_August_2019.csv")
df_dec = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_December_2019.csv")
df_feb = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_February_2019.csv")
df_jan = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_January_2019.csv")
df_july = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_July_2019.csv")
df_march = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_March_2019.csv")
df_june = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_June_2019.csv")
df_may = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_May_2019.csv")
df_nov = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_November_2019.csv")
df_oct = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_October_2019.csv")
df_sept = pd.read_csv("/content/drive/MyDrive/Data Sets for Data Visualisation/Sales_September_2019.csv")
```

```python
df_april.head()
```

```python
# Concatenate the data frames along the rows
merged_df = pd.concat([df_april, df_aug, df_dec, df_feb, df_jan, df_july, df_march, df_june, df_may, df_nov, df_oct, df_sept], ignore_index=True)

# Display the merged data frame
merged_df
```

# Cleaning the data

```python
null = merged_df.isnull()
rows_with_null = merged_df[null.any(axis=1)]
rows_with_null
```

# Add a sales column

```python
newdf['Sales'] = newdf['Quantity Ordered'] * newdf['Price Each']
newdf
```

# Converting columns to int

```python
newdf['Quantity Ordered'] = pd.to_numeric(newdf['Quantity Ordered']) # Make int
newdf['Price Each'] = pd.to_numeric(newdf['Price Each'])
```

```python
newdf.loc[:,'Quantity Ordered'] = pd.to_numeric(newdf['Quantity Ordered']) # Make int
newdf.loc[:,'Price Each'] = pd.to_numeric(newdf['Price Each'])
```

```python
newdf
```

```python
newdf.loc['Sales'] = newdf['Quantity Ordered'] * newdf['Price Each']
newdf
```

```python
# What was the best month for sales? and how much was earnerd that month?

result = newdf.groupby('Month')['Sales'].sum()
result_sorted = result.sort_values(ascending=False)
```

```python
result = newdf.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)
plt.bar(months, result['Sales'])
plt.ylabel('Sales (in Million) USD')
plt.xlabel(f"Month Number")
plt.title("Sales vs Month")
plt.xticks(months)
plt.show()
```
