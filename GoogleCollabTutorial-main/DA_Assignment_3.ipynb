# Question 5
# Normalize Data in range 0 to 1

```python
df.head()
```

```python
# Normalization
df_n = df
numeric_columns = df_n.select_dtypes(include=['float64', 'int64']).columns

for column in numeric_columns:
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())

df_n.head()
```

```python
df_s = df.sort_values(by=['track_popularity', 'acousticness'], ascending = [False, False])
df_s
```
