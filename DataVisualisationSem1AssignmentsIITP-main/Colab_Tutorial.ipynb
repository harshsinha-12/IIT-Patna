````python
# Import library and naming alias

import pandas as pd
````

## Creating a Series (1-Dimentional Data Structure)

A Series can be created using any of the following
1. Scalar Value
2. Lists
3. Dictionary


````python
# Series from Scalar value
series1 = pd.Series(10)
print(series1)
````

````python
# Series from Scalar value with defined index value
series2 = pd.Series(10, index= [0, 1 , 2 , 3])
print(series2)
````

````python
series3 = pd.Series(10, index= ['one', 'two', 'three', 'four'])
print(series3)
````

````python
# Series from list with default index
lst = [10,20,30,40,50]
series4 = pd.Series(lst)
print(series4)

print(series4.describe())
````

````python
# Series from a Dictionary
dic = {'Name':'Indrajeet','Subject':'Data-Science','Topic':'Pandas'}
series5 = pd.Series(dic)
print(series5)
````

## Creating a DataFrame (2-Dimentional Data Structure)

A data frame can be created using any of the following
1. Series
2. Lists
3. Dictionary


````python
#  Dataframe From list
lst2 = [5,18,25,30,32]
df1 = pd.DataFrame(lst2)
print(df1)
````

````python
#  Dataframe From list of lists
lst3 = [[5,18,25],[30,32]]
df2 = pd.DataFrame(lst3, index=[0,1],  columns=['two', 3, 4])
print(df2)
````

````python
#  Dataframe From Series
df3 = pd.DataFrame(series2)
print(df3)
````

````python
# DataFrame from Dictionary
dic2 = {'Sport':['Cricket','Football'],'Player':['Rohit Sharma','Sunil Chhetri']}
df4 = pd.DataFrame(dic2)
print(df4)
````

````python
# DataFrame from Dictionary of Series
name = pd.Series(['Indrajeet','Shashi'])
topic = pd.Series(['Pandas','Data'])
dic3 = {'Name':name,'Topic':topic}
df5 = pd.DataFrame(dic3)
print(df5)
````

````python
# DataFrame from List of Dictionaries
lst4 = [{'Name':'Indrajeet','Topic':'Pandas'},
        {'Name':'Shashi','Topic':'Data'}]
df6 = pd.DataFrame(lst4)
print(df6)
````

### Selection of Row or Column

1. Selection Using loc
2. Selection Using iloc
2. Selection Using [ ]

````python
# Selection using [] for column
# Syntax:- variable_name[index]

iris['SepalLengthCm']
````

````python
# Selection using [] for rows
# Syntax:- variable_name[StartRange> : StopRange - 1]
iris[0:4]
````

````python
# Selection Using loc for rows
# Syntax:- name.loc[StartRange : StopRange]
iris.loc[0:4]
````

````python
# Selection Using loc for column
# Syntax:- name.loc[index]
iris.loc[0:5,['SepalLengthCm']]
````

````python
# Selection according to some conditions

iris.loc[(iris.SepalLengthCm > 5) & (iris.SepalWidthCm < 2.5)]
````

````python
# Selection Using iloc for rows
# Syntax:- name.iloc[StartRange : StopRange - 1]
iris.iloc[0:1]
````

````python
# Selection Using iloc for column
# Syntax:- name.iloc[index]
# have to pass an integer index in the method to select a specific row/column.
# does not accept the boolean data
print(iris.iloc[0:10,0:2])
````

## To Add & Rename a column in dataframe



````python
arr = [ 10 , 20 , 30 , 40]
df_1= pd.DataFrame(arr, columns=['col1'])
df_1['New'] = 20   # new column with all same value
print(df_1)
````

````python
# Rename
df_1.columns = ['col11', 'col22']
print(df_1)
````

````python
# Rename
df_1.rename(columns = {'col22':'col2'}, inplace = True)
print(df_1)
````

````python
# Rename
df_1.set_axis(['A', 'B'], axis='columns')
print(df_1)
````

### To Delete a Column in data frame

````python
# Delete a Column Using del
df_2 = df_1.copy()
del df_2['B']
print(df_2)
````

````python
# Delete a Column Using drop()
df_3 = df_1.copy()
df_3.drop('A',axis=1,inplace=True)
print(df_3)
````

````python
# Delete a row Using drop()
df_4 = df_1.copy()
df_4.drop(index=[1,2],axis=0,inplace=True)
print(df_4)
````

````python
# Delete NULL value Using dropna()
# Delete a row Using drop()
df_5 = pd.DataFrame([[10,20],[65,70,75]])
print(df_5)
print("-------------------")
print("After deletion of NaN row")
df_6 = df_5.dropna()  # By default axis=0
print(df_6)
print("-------------------")
print("After deletion of NaN column")
df_7 = df_5.dropna(axis=1)
print(df_7)
````

### Replace Empty/Null Values

````python
# Filling value Using fillna()
df_8 = df_5.copy()
df_8.fillna(100,inplace=True)
print(df_8)
````

### Adding Values to existing DataFrame

````python
# Adding value Using loc
df_9 = df_5.copy()
df_9.loc[len(df_9.index)] = [1,None,2]
print(df_9)
````

````python
# Adding value Using append
df_10 = df_5[0:1].copy()
print(df_10)
print("-------------------")
df_11 = df_5[1:].copy()
print(df_11)
print("-------------------")
df_10 = pd.concat([df_10,df_11])
print(df_10)
````

### Replace Null Value using Mean, Median or Mode

````python
# Replacing Null with mean value
df_12 = df_9.copy()
print(df_12)
print("-------------------")
mean = df_12.loc[:,2].mean()
df_12.loc[:,2].fillna(mean, inplace=True)
print(df_12)
````

````python
# Replacing Null with mode value
df_13 = df_9.copy()
print(df_13)
print("-------------------")
mode = df_13.loc[:,2].mode()[0]
df_13.loc[:,2].fillna(mode, inplace=True)
print(df_13)
````
